1. Membuat Gamma Correction Sesuai Petunjuk

```python
original = cv.imread('/content/drive/MyDrive/Images/female.tiff')
img_rgb = cv.cvtColor(original,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
print(' Gamma Correction pada citra ')
print('-----------------------------')
try:
  gamma = int(input('Masukkan nilai Gamma: '))
except ValueError:
  print('Error, not a number')

ori = cv.cvtColor(img_rgb,cv.COLOR_BGR2RGB)
final_image = np.zeros(original.shape, original.dtype)

for y in range(ori.shape[0]):
  for x in range(ori.shape[1]):
    for c in range(ori.shape[2]):
     final_image[y,x,c] = np.clip(255 * pow(ori[y,x,c] / 255, (1 / gamma)), 0, 255)
     
final_frame = cv.hconcat((ori, final_image))
cv2_imshow(final_frame)
```

2. Membuat Simulasi Image Depth


```python
ori = cv.cvtColor(img_rgb,cv.COLOR_BGR2RGB)
final_image = np.zeros(ori.shape, ori.dtype)

print(' Simulasi Image Depth ')
print('----------------------')
try:
  bd = float(input('Masukkan nilai bit depth: '))
except ValueError:
  print('Error, not a number')

bit = 2**bd
level = 255 / (bit - 1)

for y in range(ori.shape[0]):
  for x in range(ori.shape[1]):
    for c in range(ori.shape[2]):
     final_image[y,x,c] = np.clip(round(original[y,x,c] / level) * level, 0, 255);

final_frame = cv.hconcat((ori, final_image))
cv2_imshow(final_frame)
```

3. Membuat Average Denoising 

```python
import glob
from math import log10, sqrt
```

a. Jumlah citra di average : 5

```python
original = cv.imread('/content/drive/MyDrive/Images/galaxy.jpg')
cv_img = []
for img in glob.glob('/content/drive/MyDrive/Images/noises/*.jpg'):
  n = cv.imread(img)
  cv_img.append(n)

dst = cv_img[0]
noise_number = 4 # nomor (array) gambar pada folder noises
for i in range(noise_number):
    if i == 0:
        pass
    else:
        alpha = 1.0 / (i + 1)
        beta = 1.0 - alpha
        dst = cv.addWeighted(cv_img[i], alpha, dst, beta, 0.0)

def PSNR(original, compressed): 
    mse = np.mean((original - compressed) ** 2) 
    if(mse == 0): # MSE is zero means no noise is present in the signal. 
                  # Therefore PSNR have no importance. 
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse)) 
    return psnr

compressed = dst
psnr = PSNR(original, compressed)

print('Nilai PSNR adalah',psnr,'(dB)')
galaxy_frame = cv.hconcat((original, dst))
cv2_imshow(galaxy_frame)
```
